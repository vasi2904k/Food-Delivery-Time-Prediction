# ***Libraries***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# ***Paths***

```python
INPUT_FILE = '../data/raw/Food_Delivery_Time_Prediction.csv'

OUTPUT_DIR = r'..\outputs\plots\\'

def save_plot(filename):
    path = OUTPUT_DIR + filename
    plt.savefig(path, bbox_inches='tight')
    print(f'Saved to -> {path}')
```

```python
df = pd.read_csv(INPUT_FILE)
df.head()
```

```python
sns.pairplot(df, palette='Tab10')
save_plot('pair_plot.png')
plt.show()
```

# **BoxPlots**

```python
cols = ['Distance',
       'Weather_Conditions', 'Traffic_Conditions',
       'Delivery_Person_Experience', 'Order_Priority', 'Order_Time',
       'Vehicle_Type', 'Restaurant_Rating', 'Customer_Rating', 'Delivery_Time',
       'Order_Cost', 'Tip_Amount']

for col in cols:
    plt.figure(figsize=(15,10))
    sns.boxplot(x=df[col], color = 'blue')
    plt.title(f"OUtliers in {col}")
    plt.xlabel(col)
    plt.tight_layout()
    save_plot(f'{col}_boxplot.png')
    plt.show()
```

# **Bar Plots**

```python
cols = ['Order_Priority', 'Order_Time',
       'Vehicle_Type']
for col in cols:
    plt.figure(figsize=(15,10))
    counts = df[col].value_counts()
    
    sns.barplot(x=counts.index, y=counts.values, palette='Paired')
    
    plt.title(f"{col} Distribution")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.tight_layout()
    save_plot(f'{col}_barplot.png')
    plt.show()
```

# **Count Plots**

```python
cols = ['Weather_Conditions', 'Traffic_Conditions',
        'Order_Priority', 'Order_Time',
       'Vehicle_Type']

for col in cols:
    plt.figure(figsize=(25,15))
    sns.countplot(df[col], palette='Set2')
    plt.title(f'{col} Countplot')
    plt.xlabel('counts')
    save_plot(f'{col}_countplot.png')
    plt.ylabel(f'{col}')
```
